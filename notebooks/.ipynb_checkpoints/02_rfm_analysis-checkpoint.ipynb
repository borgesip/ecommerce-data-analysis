```python
# # Anlise RFM
#
# Calcula Recncia, Frequncia e Monetrio para segmentar clientes.

# ## Importaes
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os
import seaborn as sns
import matplotlib.pyplot as plt
import sys
sys.path.append(os.path.abspath('..'))

# ## Configurao
load_dotenv()
DATABASE_URL = os.getenv("DATABASE_URL", "postgresql://postgres:senha123@localhost:5432/ecommerce_data")

# ## Extrair Dados
query = """
SELECT customerid, invoicedate, totalvalue, invoiceno
FROM sales
WHERE customerid != 'Unknown';
"""
engine = create_engine(DATABASE_URL)
df = pd.read_sql(query, engine)

# ## Calcular RFM
reference_date = df['invoicedate'].max() + pd.Timedelta(days=1)
rfm = df.groupby('customerid').agg({
    'invoicedate': lambda x: (reference_date - x.max()).days,
    'invoiceno': 'nunique',
    'totalvalue': 'sum'
}).rename(columns={
    'invoicedate': 'recency',
    'invoiceno': 'frequency',
    'totalvalue': 'monetary'
})

# ## Segmentao
# Dividir recency, frequency e monetary em quartis
rfm['R_score'] = pd.qcut(rfm['recency'], 4, labels=[4, 3, 2, 1])
rfm['F_score'] = pd.qcut(rfm['frequency'].rank(method='first'), 4, labels=[1, 2, 3, 4])
rfm['M_score'] = pd.qcut(rfm['monetary'], 4, labels=[1, 2, 3, 4])
rfm['RFM_score'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)

# Definir segmentos mais granulares
def rfm_segment(row):
    if row['RFM_score'] in ['444', '443', '434', '433']:
        return 'Campees'
    elif row['R_score'] >= 3 and row['F_score'] >= 2:
        return 'Leais'
    elif row['R_score'] <= 2 and row['F_score'] >= 3:
        return 'Promissores'
    else:
        return 'Em Risco'

rfm['Segment'] = rfm.apply(rfm_segment, axis=1)

# ## Visualizaes
# Configurar estilo
sns.set_style("whitegrid")

# Criar scatter plot com melhorias
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=rfm,
    x='recency',
    y='monetary',
    hue='Segment',
    size='frequency',
    sizes=(20, 200),  # Ajustar o intervalo de tamanhos
    alpha=0.6,  # Adicionar transparncia para reduzir sobreposio
    palette='Set2'  # Usar uma paleta de cores mais distinta
)

# No bloco de visualizao do RFM scatter plot, aps o sns.scatterplot
# Encontrar a posio mdia dos Campees
campeoes = rfm[rfm['Segment'] == 'Campees']
mean_recency_campeoes = campeoes['recency'].mean()
mean_monetary_campeoes = campeoes['monetary'].mean()

# Adicionar anotao
plt.annotate('VIP Customers', 
             xy=(mean_recency_campeoes, mean_monetary_campeoes), 
             xytext=(mean_recency_campeoes + 20, mean_monetary_campeoes * 2),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='black')

# Aplicar escala logartmica no eixo y (monetary)
plt.yscale('log')
plt.ylim(1, rfm['monetary'].max() * 1.1)  # Ajustar limite inferior para evitar zeros

# Ajustar labels e ttulo
plt.xlabel('Recncia (dias desde a ltima compra)')
plt.ylabel('Monetrio (valor total em log)')
plt.title('Segmentao RFM: Recncia vs. Monetrio (Frequncia como tamanho)')

# Ajustar legenda
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustar layout para evitar corte
plt.tight_layout()

# Salvar o grfico
plt.savefig('../visualizations/rfm_scatter_improved.png', bbox_inches='tight')
plt.show()

# Grfico de distribuio de segmentos
segment_counts = rfm['Segment'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=segment_counts.values, y=segment_counts.index, palette='Set2')
plt.title('Distribuio de Segmentos RFM')
plt.xlabel('Nmero de Clientes')
plt.ylabel('Segmento')
plt.savefig('../visualizations/rfm_segments.png')
plt.show()

# ## Insights
# Calcular mtricas para insights
total_customers = len(rfm)
total_monetary = rfm['monetary'].sum()
champions = rfm[rfm['Segment'] == 'Campees']
champions_percentage = len(champions) / total_customers * 100
champions_monetary = champions['monetary'].sum() / total_monetary * 100

print(f"Campees representam {champions_percentage:.1f}% dos clientes, mas contribuem com {champions_monetary:.1f}% da receita.")
print("Recomendaes:")
print("- Campees: Oferecer programas de fidelidade e recompensas exclusivas.")
print("- Leais: Enviar promoes personalizadas para manter engajamento.")
print("- Promissores: Incentivar compras frequentes com descontos iniciais.")
print("- Em Risco: Criar campanhas de reengajamento, como e-mails com ofertas especiais.")

# ## Explicao dos Resultados

print("\n ### Explicao dos Resultados da Anlise RFM")
print("- **Distribuio dos Segmentos**:")
print("  - O maior segmento  'Em Risco' (~1.750 clientes), indicando que muitos clientes no compram h muito tempo ou tm baixa frequncia e valor de compra. Isso sugere um alto risco de churn e a necessidade de estratgias de reengajamento.")
print("  - 'Leais' (~1.500 clientes) e 'Campees' (~1.000 clientes) formam uma base slida de clientes engajados, que compram com frequncia e recentemente.")
print("  - 'Promissores' (~750 clientes) so o menor grupo, mas mostram potencial para se tornarem mais engajados com incentivos adequados.")
print("")
print("- **Contribuio de Receita**:")
print(f"  - Apesar de representarem apenas {champions_percentage:.1f}% dos clientes, os 'Campees' contribuem com {champions_monetary:.1f}% da receita, destacando sua importncia para o negcio.")
print("  - Isso indica que focar em reter esses clientes de alto valor pode maximizar a receita a longo prazo.")
print("")
print("- **Padres Visuais**:")
print("  - No scatter plot, os 'Campees' esto concentrados na regio de baixa recncia (0-50 dias) e alto valor monetrio (10^4 a 10^5), confirmando que so clientes recentes e de alto valor.")
print("  - Os 'Em Risco' esto espalhados por toda a faixa de recncia, mas principalmente com valores monetrios mais baixos (10^1 a 10^3), indicando desengajamento.")
```
